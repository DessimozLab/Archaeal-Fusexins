```python
import pandas as pd
header = 'targetname accession queryid accession2    E-value  score  bias   domain-E-value  domain-score  domain-bias   exp reg clu  ov env dom rep inc description'.split()
def hmmer2df(filename):
    framedict = {}
    with open(filename , 'r') as infile:
        for i,line in enumerate(infile):
            if line[0] != '#':
                words = line.split()
                words[18] = ''.join([ w + ' ' for w in words[18:-1]])
                #print(words)
                framedict[i] = dict(zip( header , words[0:19]))
        df = pd.DataFrame.from_dict(framedict , orient='index')
        numerical = 'E-value  score  bias   domain-E-value  domain-score  domain-bias   exp reg clu  ov env dom rep inc'.split()
        df[   numerical ] = df[numerical].astype(float)
        return df
```

```python
hmmerfile = './hmmer_vs_metaclust/fsxa_hmmersearch.tab'
hmmerdf = hmmer2df(hmmerfile)
print(hmmerdf.head())
```

```python
hmmerdf.hist( figsize=( 20 , 20))
```

```python
sub = hmmerdf[ hmmerdf['E-value'] < 10**-4 ]
sub = sub[ sub['env'] < 2 ]
print(sub)
```
