# Overview
Pairwise similarity of gene content among mobile elements is measured by means of Jaccard index computation. Matches between ORFs are defined based on homologous groups, and an standard Jaccard index is computed.

```python
# import libraries
import os
import glob
import itertools
from Bio import SeqIO
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# set figure dimensions inside Jupyter notebook
plt.rcParams['figure.figsize'] = [20, 20]
```

```python
# create dictionary between ORF ID types
orf_annotation = pd.read_csv('../results/MEs_predicted_orfs_renamed/data/annotation_table_ORFs_renamed.tsv',  sep = '\t')
orf_annotation['Genome ID'] = orf_annotation['Genome ID'].astype(str)

getorfid2finalanot_dict = {}
for getorf_id, code_id in zip(orf_annotation['getorf featured ID'].to_list(), orf_annotation['ORF ID'].to_list()):
    getorfid2finalanot_dict.update({getorf_id.split(' ')[0]: code_id})
```

```python
featureid2getorfid_dict = {}
[featureid2getorfid_dict.update({feature_id: getorf_id}) for feature_id, getorf_id in zip(orf_annotation['Featured ID in Genome Annotation'][pd.notnull(orf_annotation['Featured ID in Genome Annotation'])].to_list(),
                                                                                          orf_annotation['getorf featured ID'][pd.notnull(orf_annotation['Featured ID in Genome Annotation'])].to_list())]

for key in list(featureid2getorfid_dict.keys()):
  try:
    featureid2getorfid_dict[key] = featureid2getorfid_dict[key].split(' ')[0]
  except:
    None
```

```python
# define function to perform Jaccard index computation between two sets of genes based on homolog clusters
def compute_jaccard(gene_set_1, gene_set_2, homolog_groups_dir):  
  # getting gene lists
  gene_list_1 = gene_set_1[1]
  gene_list_2 = gene_set_2[1]
  # tagging all elements of each set with the species name
  #gene_list_1_tag = [(gene_set_1[0], getorfid2finalanot_dict[gene]) for gene in gene_list_1]
  gene_list_1_tag = [(gene_set_1[0], gene) for gene in gene_list_1]
  #gene_list_2_tag = [(gene_set_2[0], getorfid2finalanot_dict[gene]) for gene in gene_list_2]
  gene_list_2_tag = [(gene_set_2[0], gene) for gene in gene_list_2]
  # get union set
  union_gene_set = list(set(gene_list_1_tag+gene_list_2_tag))
  # get homolog clusters components and store these in a list
  homologs_sets = []
  for homolog_group_file in glob.glob(homolog_groups_dir):
    # load genes
    genes_in_group = [record.id.split(' ')[0] for record in SeqIO.parse(homolog_group_file, 'fasta')]
    # add them as set in list of homologs
    homologs_sets.append(set(genes_in_group))
  # get intersection based on homology matrix
  intersection = 0
  for pairwise_comp in itertools.combinations(union_gene_set, 2):
    # depack variables
    gene_set_1, gene_set_2 = pairwise_comp
    # compare if genes are from different gene_set
    if gene_set_1[0] != gene_set_2[0]:
      # check if both elements belong to a same homology group, append True to <coincidence_in_cluster>
      coincidence_in_cluster = []
      for homolog_set in homologs_sets:
        if (gene_set_1[1] in homolog_set) & (gene_set_2[1] in homolog_set):
          coincidence_in_cluster.append(True)
      # else appending False
        else:
          coincidence_in_cluster.append(False)
      # if they belong to any cluster at the same time, append 1 to element_coincidences; else append zero
      if any(coincidence_in_cluster):
        intersection = intersection+1
      else:
        intersection = intersection
    # return Jaccard index value
  return "%.9f" % (intersection/len(union_gene_set))
```

```python
# create dictionary from id2genome
genomes_and_taxonomy = pd.read_csv('../data/genomes_and_taxonomy.csv', 
                                    sep = ',')
genomes_and_taxonomy['Genome_id'] = genomes_and_taxonomy['Genome_id'].astype(str)

id2genome_dict = {}
for id, taxon in zip(genomes_and_taxonomy.Genome_id.to_list(), genomes_and_taxonomy.Taxonomy.to_list()):
  # update dictionary
  id2genome_dict.update({id.replace('GCA_', ''): taxon})
```

```python
# list all mobile elements ORF files
me_orf_sets = [(id2genome_dict[orf_file.split('/')[3].replace('.predicted_orfs.faa', '').replace('ME.', '').replace('.MAG', '').replace('.metagenome', '')],
                orf_file) for orf_file in glob.glob('../results/MEs_predicted_orfs/*faa') if os.path.exists(orf_file.replace('MEs_predicted_orfs', 'ME_PATRIC_annotated_features').replace('.predicted_orfs.faa', '.PATRIC_annotated_features.faa').replace('.MAG', '').replace('.metagenome', ''))]

# get list of target organisms
mobile_element_lengths_table_filtered = pd.read_csv('../results/target_organisms.tsv', sep ='\t')

# filter me_orf_sets to contemplate only target organisms
me_orf_sets_filtered = []
for me_set in me_orf_sets:
    # depack variables
    taxonomy, me_file = me_set
    if taxonomy in mobile_element_lengths_table_filtered['Taxonomy'].to_list():
        me_orf_sets_filtered.append(me_set)
        
# get variables linked again
me_orf_sets = me_orf_sets_filtered

# create homology matrix based on homologous groups (i.e. 1 if pair of genes is homologous, 0 if not)
# create dictionary with organisms and ORFs in homolog sets
## first a dictionary from ORFs to organisms
taxon2orfid_dict = {}
for sets in orf_annotation.groupby('Genome ID'):
  genome = sets[0]
  taxon = id2genome_dict[genome]
  table = sets[1]
  seqids = [seqid.split(' ')[0] for seqid in table['getorf featured ID'].to_list()]
  taxon2orfid_dict.update({taxon: seqids})
```

```python
seqs_in_homolog_groups = []
for group in glob.glob('../results/trying_to_catch_homologs/results/connected_groups_filtered/fastas/protein/*faa'):
  # load seqs in groups
  seqs = [record.id.split(' ')[0] for record in SeqIO.parse(group, 'fasta')]
  # add them to list
  seqs_in_homolog_groups.extend(seqs)
  
org2orf_in_homolog_groups_dict = {}
for taxon in list(taxon2orfid_dict.keys()):
  seqs_of_taxon = []
  for seq in taxon2orfid_dict[taxon]:
    if seq in seqs_in_homolog_groups:
      seqs_of_taxon.append(seq)
  org2orf_in_homolog_groups_dict.update({taxon: seqs_of_taxon})
```

```python
# get gene_sets in format (<ME. species in taxonomy>, [<gene list with record.id>])
gene_sets_mobile_elements = []
for gene_set in me_orf_sets:
  # depack variables
  taxon, orf_file = gene_set
  # first get annotated features in ME
  annotated_file = orf_file.replace('MEs_predicted_orfs', 'ME_PATRIC_annotated_features').replace('.predicted_orfs.faa', '.PATRIC_annotated_features.faa').replace('.MAG', '').replace('.metagenome', '')
  patric_annotated = []
  for record in SeqIO.parse(annotated_file, 'fasta'):
    try:
      patric_annotated.append(featureid2getorfid_dict[record.id])
    except:
      None
  # aggregate ORFs in homologous groups
  seqs_in_homolog_groups = org2orf_in_homolog_groups_dict[taxon]
  final_set = list(set(patric_annotated + seqs_in_homolog_groups))
  # append records
  gene_sets_mobile_elements.append((taxon, final_set))
  
# iterate over all non-redundant pairwise combinations of species
import math
rows = [sets[0] for sets in gene_sets_mobile_elements]
columns = [sets[0] for sets in gene_sets_mobile_elements]
jaccard_indexes = []
```

```python
# creating heatmap
import seaborn as sns; sns.set_theme(color_codes=True)
```

```python
# performing Jaccard index calculation with homolog sets instead of genes
# create auxiliary function
def get_homolog_groups_of_genes(gene_list):
  homologous_groups = []
  for gene in gene_list:
    # if a gene is in a homologous set, then add the set to <homologous_groups_1>
    for homolog_group in list(homolog_groups_dict.keys()):
      # get sequences in homolog group
      homolog_group_seqs = homolog_groups_dict[homolog_group]
      if gene in homolog_group_seqs:
        homologous_groups.append(homolog_group)
  return homologous_groups

homolog_groups_dict = {}
for group in glob.glob('../results/trying_to_catch_homologs/results/connected_groups_filtered/fastas/protein/*faa'):
  # load seqs in groups
  seqs = [record.id.split(' ')[0] for record in SeqIO.parse(group, 'fasta')]
  # add them to list
  group_name = group.split('/')[7]
  homolog_groups_dict.update({group_name: seqs})

# create function to perform Jaccard with homolog sets and not with genes
def compute_jaccard_homologsets(gene_set_1, gene_set_2, homolog_groups_dir):    
  # getting gene lists
  gene_list_1 = gene_set_1[1]
  gene_list_2 = gene_set_2[1]
  # getting homologous groups to where this genes belong
  # create dictionary with homologous groups and their sequences
  # running the auxiliary function over the two gene lists
  homolog_groups_1 = get_homolog_groups_of_genes(gene_list_1)
  homolog_groups_2 = get_homolog_groups_of_genes(gene_list_2)
  # get union set
  union_homologs_set = list(set(homolog_groups_1+homolog_groups_2))
  # get intersection
  intersection_homologs_set = list(set(homolog_groups_1) & set(homolog_groups_2))
  # calculate and return Jaccard index value
  try:
    return "%.9f" % (len(intersection_homologs_set)/len(union_homologs_set))
  except Exception:
    return 0

# running this function and plotting
rows = [sets[0] for sets in gene_sets_mobile_elements]
columns = [sets[0] for sets in gene_sets_mobile_elements]
jaccard_indexes_homolog_groups = []
for x in itertools.product(gene_sets_mobile_elements, gene_sets_mobile_elements):
  # depack variables
  me_orf_set_1, me_orf_set_2 = x
  # perform jaccard index calculation and append it to <jaccard_indexes> if not same species
  #if me_orf_set_1[0] != me_orf_set_2[0]:
  jaccard_index = compute_jaccard_homologsets(gene_set_1 = me_orf_set_1, 
                                  gene_set_2 = me_orf_set_2, 
                                  homolog_groups_dir = '../results/trying_to_catch_homologs/results/connected_groups_filtered/fastas/protein/*faa')
  # append value
  jaccard_indexes_homolog_groups.append(float(jaccard_index))
 
```

```python
jaccard_indexes_array_homolog_groups = np.array(jaccard_indexes_homolog_groups)
jaccard_indexes_table_homolog_groups = pd.DataFrame(jaccard_indexes_array_homolog_groups.reshape(-1, len(columns)), index=rows, columns=columns)
```
